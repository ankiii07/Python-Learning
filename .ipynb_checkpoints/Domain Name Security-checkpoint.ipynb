```python
def draw_histograms(dataframe, features, rows, cols):
    fig=plt.figure(figsize=(20,20))
    for i, feature in enumerate(features):
        ax=fig.add_subplot(rows,cols,i+1)
        dataframe[feature].hist(bins=20,ax=ax,facecolor='midnightblue')
        ax.set_title(feature+" Distribution",color='DarkRed')
        
    fig.tight_layout()  
    plt.show()
draw_histograms(df,df.columns,6,3)
```

```python
df.Action.value_counts()
```

```python
categorical_val = []
continous_val = []
for column in df.columns:
    print('--------------------------------------------------------------')
    print(f"{column} : {df[column].unique()}")
    if len(df[column].unique()) <= 10:
        categorical_val.append(column)
    else:
        continous_val.append(column)
```

```python
Action = {'allow':1,'deny':0,'drop':2,'reset-both':3}
df['Action'] = df.Action.map(Action)
```

```python
df.head()
```

```python
plt.figure(figsize=(15, 15))

for i, column in enumerate(categorical_val, 1):
    plt.subplot(3, 3, i)
    df[df["Action"] == 0][column].hist(bins=40, color='blue', label='access = NO', alpha=1,width=0.2)
    df[df["Action"] == 1][column].hist(bins=40, color='red', label='access = YES', alpha=1,width=0.2)
    df[df["Action"] == 2][column].hist(bins=40, color='yellow', label='access = maybe', alpha=1,width=0.2)
    df[df["Action"] == 3][column].hist(bins=40, color='green', label='access = reset', alpha=1,width=0.2)
    plt.legend()
    plt.xlabel(column)
```
