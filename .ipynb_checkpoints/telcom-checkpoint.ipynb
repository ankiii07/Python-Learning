```python
categorical_val = []
continous_val = []
for column in df.columns:
    print('--------------------------------------------------------------')
    print(f"{column} : {df[column].unique()}")
    if len(df[column].unique()) <= 10:
        categorical_val.append(column)
    else:
        continous_val.append(column)
```

```python
def draw_histograms(dataframe, features, rows, cols):
    fig=plt.figure(figsize=(20,20))
    for i, feature in enumerate(features):
        ax=fig.add_subplot(rows,cols,i+1)
        dataframe[feature].hist(bins=20,ax=ax,facecolor='blue')
        ax.set_title(feature+" Distribution",color='Red')
        
    fig.tight_layout()  
    plt.show()
draw_histograms(df,df.columns,6,3)
```

```python
plt.figure(figsize=(15, 15))

for i, column in enumerate(categorical_val, 1):
    plt.subplot(4, 4, i)
    df[df["Churn"] == 0][column].hist(bins=40, color='blue', label='business does = NO', alpha=1,width=0.2)
    df[df["Churn"] == 1][column].hist(bins=40, color='red', label='business does = YES', alpha=1,width=0.2)
    plt.legend()
    plt.xlabel(column)
```
