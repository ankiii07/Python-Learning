```python
from sklearn.model_selection import train_test_split
```

```python
len(train_x)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
```

```python
model.fit(train_x,train_y)
```

```python
model.score(test_x,test_y)
```
