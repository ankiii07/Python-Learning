```python
import matplotlib.pyplot as plt
```

## Scatter Plot

```python
x = [1,2,3,4,5,6,7,8,9]
y=[4,3,7,5,3,2,5,7,6]
 
plt.scatter(x,y,label = 'skitscat', color = 'k')
plt.xlabel('x')
plt.ylabel('y')
plt.title('graph study')
plt.legend()
```

```python
x = [1,2,3,4,5,6,7,8,9]
y=[4,3,7,5,3,2,5,7,6]
 
plt.scatter(x,y,label = 'skitscat', color = 'r', marker = '*', s =100)
plt.xlabel('x')
plt.ylabel('y')
plt.title('graph study')
plt.legend()
```

## Stack Plot

```python
days = [1,2,3,4]

sleeping=[4,7,2,8]
eating = [5,3,6,9]
working =[2,9,5,4]
playing =[5,4,8,7]

plt.stackplot(days,sleeping,eating,working,playing, colors =['m','r','k','c'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('graph study')
plt.legend()
plt.show()
```

```python
days = [1,2,3,4]

sleeping=[4,7,2,8]
eating = [5,3,6,9]
working =[2,9,5,4]
playing =[5,4,8,7]

plt.plot([],[],color ='m', label='sleeping')
plt.plot([],[],color ='r', label='eating')
plt.plot([],[],color ='k', label='working')
plt.plot([],[],color ='c', label='playing')

plt.stackplot(days,sleeping,eating,working,playing, colors =['m','r','k','c'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('graph study')
plt.legend()
plt.show()
```

## Pie Chart

```python
days = [1,2,3,4]

sleeping=[4,7,2,8]
eating = [5,3,6,9]
working =[2,9,5,4]
playing =[5,4,8,7]

slices = [15,8,5,10]
activities = ['sleeping','eating','working','playing']
cols =['m','r','k','c']

plt.pie(slices,
        labels=activities,
        colors = cols,
        startangle =90, 
        shadow =True,
        explode = (0.0,0.1,0.0,0.2),
       )

plt.xlabel('x')
plt.ylabel('y')
plt.title('graph study')
plt.legend()
plt.show()
```
