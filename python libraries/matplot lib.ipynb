# matplot library

```python
import matplotlib.pyplot as plt
```

```python
days=[1,2,3,4,5,6,7]
views=[123,354,234,678,645,876,986]
```

```python
plt.plot(days,views)
```

```python
plt.xlabel("no. of days")
plt.ylabel("no. of views")
plt.title("social media views")
plt.plot(days,views)
```

```python
plt.xlabel("no. of days")
plt.ylabel("no. of views")
plt.title("social media views")
plt.plot(days,views,color="red",linewidth="5",linestyle="--")
```

```python
plt.xlabel("no. of days")
plt.ylabel("no. of views")
plt.title("social media views")
plt.plot(days,views,color="red",linewidth="5",linestyle="--",marker="o",markerfacecolor="black")
```
